```python
import numpy as np


#formatted the matrix data to numpy array
counts = np.array([
    [8, 12, 21, 9, 4, 2, 21, 21, 3, 10, 8, 5, 7, 25, 4, 2, 2, 25], # A
    [7, 4, 1, 6, 2, 3, 3, 3, 1, 0, 2, 0, 7, 0, 3, 3, 24, 0], # C
    [3, 2, 1, 8, 2, 21, 2, 2, 0, 1, 0, 1, 0, 1, 0, 15, 0, 2], # G
    [9, 9, 4, 4, 19, 1, 1, 1, 23, 16, 17, 21, 13, 1, 20, 7, 1, 0] # T
])


total_sites = 27

pseudo_counts = counts + 1


adjusted_sites = total_sites + 4 # Adding 1 for each nucleotide (A, C, G, T)


F_bj = counts / total_sites

F_prime_bj = pseudo_counts / adjusted_sites


F_bo = 0.25


W_bj = np.log2(F_prime_bj / F_bo)

W_bj # Position Specific Weight Matrix
```

```python
import bz2
import os


file_path = '/content/E_coli_K12_MG1655.400_50.bz2'

# Decompressing  the BZ2 file
decompressed_data = bz2.BZ2File(file_path, 'r').read()


decompressed_file_path = 'E_coli_K12_MG1655.400_50.txt'
with open(decompressed_file_path, 'wb') as f:
    f.write(decompressed_data)


d_text = decompressed_data.decode('utf-8')
print("\n".join(d_text.split('\n')[:10]))
```

```python
import re

# Convert the decompressed text into a list of lines
lines = d_text.strip().split('\n')

# Initialize a dictionary to store gene IDs and sequences
gene_sequences = {}

# Regular expression to parse the lines
line_regex = re.compile(r'^(\d+)\s*\\(.*)$')

# Parse the file to extract gene IDs and sequences
for line in lines:
    match = line_regex.match(line)
    if match:
        gene_id = match.group(1).strip()
        sequence = match.group(2).strip()
        gene_sequences[gene_id] = sequence

# Check the first few entries to confirm parsing
list(gene_sequences.items())[:5]
```

```python

def compute_score(sequence, weight_matrix):
    max_score = float('-inf')
    sequence_length = len(sequence)
    window_size = weight_matrix.shape[1]
    nucleotide_map = {'A': 0, 'C': 1, 'G': 2, 'T': 3}

    # Slide the weight matrix over the sequence
    for start in range(sequence_length - window_size + 1):
        current_score = 0
        for j in range(window_size):
            nucleotide = sequence[start + j]
            if nucleotide in nucleotide_map:
                current_score += weight_matrix[nucleotide_map[nucleotide], j]
        max_score = max(max_score, current_score)

    return max_score

# Dictionary to store the maximum score for each gene
gene_scores = {}

# Computing the  scores for each gene's sequence
for gene_id, sequence in gene_sequences.items():
    gene_scores[gene_id] = compute_score(sequence, W_bj)

# Find the top 30 gene IDs based on their scores
top_gene_ids = sorted(gene_scores, key=gene_scores.get, reverse=True)[:30]



# Prepare a detailed output for the top gene IDs with their sequence and positions
gene_details = [(gene_id, gene_sequences[gene_id]) for gene_id in top_gene_ids]


gene_details
```
